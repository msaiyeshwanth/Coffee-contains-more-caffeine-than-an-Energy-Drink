```python
import pandas as pd
from transformers import pipeline
```

```python
text = """#RRRMovie is an invigorating, indelible EXPERIENCE. S.S. Rajamouli & Co. deliver.
Huge spectacle-driven action sequences are outrageous, audacious & entertaining.
Drama, romance & heroics have real stakes & satisfying thrills. Fantastic dance & fight choreography.
BONKERS great."""
```

```python
classifier = pipeline("text-classification")
outputs = classifier(text)
pd.DataFrame(outputs)
```

```python
ner_tagger = pipeline("ner", aggregation_strategy="simple")
outputs = ner_tagger(text)
pd.DataFrame(outputs)
```

```python
reader = pipeline("question-answering")
question = "what is the review about?"
outputs = reader(question=question, context=text)
pd.DataFrame([outputs])
```

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=50, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```

```python
translator = pipeline("translation_en_to_de",
                      model="Helsinki-NLP/opus-mt-en-de")
outputs = translator(text, clean_up_tokenization_spaces=True, min_length=50)
print(outputs[0]['translation_text'])
```

```python
from transformers import set_seed
set_seed(42)
generator = pipeline("text-generation")
response = "I am excited to hear stunning response for RRR."
prompt = text + "\n\n My response:\n" + response
outputs = generator(prompt, max_length=200)
print(outputs[0]['generated_text'])
```
